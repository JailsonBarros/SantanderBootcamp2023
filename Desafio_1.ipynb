# Desafio 1 (ETL com Python)

**Contexto:** Você é um cientista de dados em uma Sorveteria famosa e recebeu a tarefa de envolver analisar dados de venda com o objetivo de identificar tendencias de vendas e oportunidades de melhorias.

**Condições do Problema:**

1. Você recebeu uma planilha simples, em formato CSV ('vendas.csv'), com dados das vendas.
2. Seu trabalho é carregar os dados desse arquivo CSV.
3. Depois de obter os dados das vendas, você irá processar os dados e analisar.
4. Uma vez que o processamento e analise dos dados estiver pronta, você irá gerar arquivos CSV com dados releventes.



## **E**xtract

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import re

df_vendas = pd.read_csv("vendas.csv")
df_vendas
```

## **T**ransform

```python
def extract_price(price_str):
    price = re.sub(r'[^\d.,]', '', price_str)
    return float(price.replace(',', '.'))

df_vendas['preco'] = df_vendas['preco'].apply(extract_price)
df_vendas['total_venda'] = df_vendas['quantidade'] * df_vendas['preco']

analise_por_produto = df_vendas.groupby('produto')['total_venda'].sum()
analise_por_genero = df_vendas.groupby('genero')['total_venda'].sum()
```

## **L**oad

```python
df_vendas.to_csv('dados_de_vendas.csv', index=False)
analise_por_produto.to_csv('analise_por_produto.csv')
analise_por_genero.to_csv('analise_por_genero.csv')
```
